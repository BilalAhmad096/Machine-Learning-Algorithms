```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Read the Excel file
data = pd.read_csv('Housing.csv') # put the filename with extension here if file kept in same location otherwise provide full address.
```

```python
# Convert categorical variables to dummy/indicator variables
data = pd.get_dummies(data, columns=['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus'], drop_first=True)
```

```python
# Define features and target variable
X = data.drop('price', axis=1)
y = data['price']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# random state is kept same throughout the code so that same rows and used for training and testing.
# test size can be changed, 0.3 means 30% is for testing and 70% for training.
```

```python
ridge_model = Ridge(alpha=1.0)  # Alpha is the regularization parameter
scores = cross_val_score(ridge_model, X, y, cv=5)
print("Cross-validation scores:", scores)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
# Make predictions
predictions = ridge_model.predict(X_test)
```

```python
array1 = y_test.to_numpy()
```

```python
# plot the data
plt.plot(array1, label='Actual')
plt.plot(predictions, label='Predicted')
plt.legend()
plt.xlabel('')
plt.ylabel('')
plt.title('')
#plt.savefig('', dpi = 600) # dpi is used to control the quality of image.
#plt.savefig('plot.pdf')
plt.show()
```
