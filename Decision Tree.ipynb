```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Read the Excel file
data = pd.read_csv('Housing.csv') # put the filename with extension here if file kept in same location otherwise provide full address.
```

```python
# Convert categorical variables to dummy/indicator variables
data = pd.get_dummies(data, columns=['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus'], drop_first=True)
```

```python
# Define features and target variable
X = data.drop('price', axis=1)
y = data['price']
```

```python
# Split data into features and target variable
X = data[['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
    'parking', 'prefarea', 'furnishingstatus']].values # Put the column names here which are the features of the dataset.
y = data[['price']].values # Put the column names here which are the target value of the dataset.
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# random state is kept same throughout the code so that same rows and used for training and testing.
# test size can be changed, 0.3 means 30% is for testing and 70% for training.
```

```python
# Create Decision Tree Regressor object
dt = DecisionTreeRegressor(random_state=42)
```

```python
# Train the model using the training sets
dt.fit(X_train, y_train)
```

```python
# Make predictions using the testing set
y_pred = dt.predict(X_test)
```

```python
array1 = y_test.to_numpy()
```
